Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend.

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
schools['percent_tested'] = schools['percent_tested'].fillna(schools['percent_tested'].mean())
```

```python
math_score_cutoff = 0.8*800
math_score_cutoff
```

```python
best_math_schools = schools[['school_name','average_math']].sort_values('average_math', ascending=False)
best_math_schools = best_math_schools[best_math_schools['average_math'] >= math_score_cutoff]
best_math_schools
```

```python
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
top_10_schools = schools[['school_name','total_SAT']].sort_values('total_SAT', ascending=False).head(10)
top_10_schools
```

```python
schools
```

```python
average_SAT = schools['total_SAT'].mean()
```

```python
# Group by borough and calculate the standard deviation of total_SAT
borough_stats = schools.groupby('borough')['total_SAT'].agg(['count', 'mean', 'std'])

# Rename columns for clarity
borough_stats = borough_stats.rename(columns={
    'count': 'num_schools',
    'mean': 'average_SAT',
    'std': 'std_SAT'
})

# Find the borough with the largest standard deviation
largest_std_dev_borough = borough_stats['std_SAT'].idxmax()

# Create the largest_std_dev DataFrame
largest_std_dev = pd.DataFrame({
    'borough': [largest_std_dev_borough],
    'num_schools': [borough_stats.loc[largest_std_dev_borough, 'num_schools']],
    'average_SAT': [borough_stats.loc[largest_std_dev_borough, 'average_SAT']],
    'std_SAT': [borough_stats.loc[largest_std_dev_borough, 'std_SAT']]
})

# Round numeric values to two decimal places
numeric_cols = ['num_schools', 'average_SAT', 'std_SAT']
largest_std_dev[numeric_cols] = largest_std_dev[numeric_cols].round(2)

largest_std_dev
```
